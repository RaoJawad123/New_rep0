```python
# Import liberaries
import seaborn as sns
import pandas as pd
import numpy as np
#canvas (Baloon Board)
sns.set(style='whitegrid')

# Load data
kashti = seaborn.load_dataset("titanic")
kashti.head()
```

```python
sns.boxplot(x="survived", y="age", data=kashti)
```

```python
p1 = sns.boxplot(x="survived", y="age", showmeans = True, data=kashti)
```

```python
import matplotlib.pyplot as plt
p1 = sns.boxplot(x="survived", y="age", showmeans = True,
                 meanprops = {"marker":"*", "markersize":"12", "markeredgecolor":"red"}, data=kashti)
#show Labels
plt.xlabel('How many survived', size = 14),
plt.ylabel("Age (years)", size = 16),
plt.title("Titanic k dobey te bachey", size = 16, weight='bold')
plt.show()
```
