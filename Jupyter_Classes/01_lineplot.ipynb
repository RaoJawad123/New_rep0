```python
# Import liberaries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y = "sepal_width", data =phool)
plt.show()
```

```python
# Adding title

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y = "sepal_width", data =phool)
plt.title("Phoolo ka plot")
plt.show()
```

```python
# Adding Limits

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y = "sepal_width", data =phool)
plt.title("Phoolo ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# Adding styles

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y = "sepal_width", data =phool)
plt.title("Phoolo ka plot")
sns.set_style("dark")
plt.show()
```

## Size of figure

```python
# changing figure size

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

plt.figure(figsize=(7,5))
# draw a line plot
sns.lineplot(x="sepal_length", y = "sepal_width", data =phool)
plt.title("Phoolo ka plot")
sns.set_style("dark")
plt.show()
```
